```python
# Ex: Random Walk
# Option 1
def random_walk(n):
    """Return coordinates after 'n' blocks random walk."""
    x = 0
    y = 0
    for i in range(n):
        step = random.choice(['N', 'S', 'E', 'W'])
        if step == 'N':
            y += 1
        elif step == 'S':
            y -= 1
        elif step == 'E':
            x += 1
        else:
            x -= 1
    return (x, y)

for i in range(25):
    walk = random_walk(10)
    print(walk, "Distance from home = ", abs(walk[0]) + abs(walk[1]))
```

```python
# Ex: Random Walk
# Option 2
def random_walk_2(n):
    """Return coordinates after 'n' blocks random walk."""
    x, y = 0, 0
    for i in range(n):
        (dx, dy) = random.choice([(0, 1), (0, -1), (1, 0), (-1, 0)]) # dx and dy are difference in x and difference in y
        x += dx
        y += dy
    return (x, y)

for i in range(25):
    walk = random_walk(10)
    print(walk, "Distance from home = ", abs(walk[0]) + abs(walk[1]))
```

```python
# Monte Carlo Simulation for Random Walk
number_of_walks = 10000

for walk_length in range (1, 31):
    no_transport = 0 # Number of walk 4 or fewer blocks from home
    for i in range(number_of_walks):
        (x, y) = random_walk_2(walk_length)
        distance = abs(x) + abs(y)
        if distance <= 4:
            no_transport += 1
    no_transport_percentage = float(no_transport) / number_of_walks
    print("Walk size =", walk_length, " / % of no transport = ", 100*no_transport_percentage)
```

```python
coin = np.random.randint(0, 2) # Randomly generates 0 or 1
print(coin)
if coin == 0:
    print("head")
else:
    print("tails")

outcomes = []
for x in range(10):
    coin = np.random.randint(0, 2)
    if coin == 0:
        outcomes.append("heads")
    else:
        outcomes.append("tails")    
print(outcomes) # NOT a random walk. Just a list of outcomes

# Random Walk
tails = [0]
for x in range(10):
    coin = np.random.randint(0, 2)
    tails.append(tails[x] + coin)
print(tails) # Increments by one when tails is thrown. Last number is number of tails.

final_tails = []
for x in range(10000):
    tails = [0]
    for x in range(10):
        coin = np.random.randint(0, 2)
        tails.append(tails[x] + coin)
    final_tails.append(tails[-1])
#print(final_tails) # prints a list of number of times tails occurred.
# Makes a histogram
plt.hist(final_tails, bins = 10)
plt.show()
```

```python
# Empire State Building
np.random.seed(123)
step = 50

dice = np.random.randint(1, 7)

if dice <= 2:
    step -= 1
elif dice > 2 and dice < 6:
    step += 1
else:
    step = step + np.random.randint(1, 7)

#print(dice)
#print(step)
plt.clf()
all_walks = []

for i in range(1000):
    random_walk = [0]
    for x in range(100):
        step = random_walk[-1]
        dice = np.random.randint(1, 7)
        if dice <= 2:
            step = max(0, step - 1) # This ensures that step cannot be less than 0.
            # step -= 1
        elif dice <= 5:
            step += 1
        else:
            step = step + np.random.randint(1, 7)
        if np.random.rand() <= 0.005: # To show where I fell down (0.5% chance)
            step = 0
        random_walk.append(step) # Creates a list of steps with the last index being total number of steps
    # print(random_walk)
    all_walks.append(random_walk) # Creates a list of all random walks
#print(all_walks)

# Convert all_walks to NumPy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
#plt.show()

# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()

# Visualize the random walk
#plt.plot(random_walk)
#plt.show()
```
