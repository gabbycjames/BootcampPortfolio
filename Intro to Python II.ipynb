```python
data = pd.read_csv('countries.csv')
```

```python
data
```

```python
# Compare the population growth in the US and China
us = data[data.country == 'United States'] # This list only includes where data is True aka is US
china = data[data.country == 'China']
plt.plot(us.year, us.population / (10 ** 6))
plt.plot(china.year, china.population / (10**6))
plt.legend(['United States', 'China'])
plt.ylabel('population')
plt.xlabel('year')
plt.show()
```

```python
# To show percentage growth from the first year
# Will show growth us.population / us.population.iloc[0] * 100

plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.legend(['United States', 'China'])
plt.ylabel('population growth (first year = 100)')
plt.xlabel('year')
plt.show()
```

```python
# Dictionaries
# Keys must be strings or numbers but values can be any type
d = {} # Option 1
d = dict() # Option 2
d["George"] = 24 # To add a key value pair to a dictionary
d["Tom"] = 32
d["Jenny"] = 16
print(d["George"])
```

```python
# How to iterate over key value pairs in dictionaries in python
for key, value in d.items():
    print("key:")
    print(key)
    print("value:")
    print(value)
    print(" ")
```
