```python
# Import things
import matplotlib.pyplot as plt

%matplotlib inline

x = list(range(0, 10))
y = list(range(-10, 0))

plt.plot(x, y)
```

```python
a = [0, -100, 25, 67, -323]
b = [0, 3, 7, 3, 9]
plt.plot(a, b)
```

```python
plt.axis([-50, 80, 2, 8]) #to show a specific portion of the graph
plt.plot(a, b)
```

```python
# To add a label to the graph
plt.title("Triangle")
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.axis([-50, 80, 2, 8]) #to show a specific portion of the graph
plt.plot(a, b)
```

```python
# To change the x-axis ticks to spell Hello
plt.title("Triangle")
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.axis([-50, 80, 2, 8]) #to show a specific portion of the graph
plt.xticks((-40, -20, 0, 20, 40, 60, 80), ('h', 'e', 'l', 'l', 'o', 'o', 'o')) #here is where we change axis
plt.plot(a, b)
```

```python
# To change the color of the line
plt.title("Triangle")
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.axis([-50, 80, 2, 8]) #to show a specific portion of the graph
plt.xticks((-40, -20, 0, 20, 40, 60, 80), ('h', 'e', 'l', 'l', 'o', 'o', 'o')) #here is where we change axis
plt.plot(a, b, color = "red") #here is where I change the color
```

```python
# To create a Histogram
plt.hist(a)
```

```python
# To show a graph with two pieces x and y
plt.hist(x)
plt.hist(y)
```

```python
# To make a graph of world population
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)
plt.show() #to display the plot
```

```python
# To make a scatter plot with the same data
plt.scatter(year, pop)
plt.show()
```

```python
# To create a histogram
values = [0, 0.6, 1.4, 1.6, 2.2, 2.5, 2.6, 3.2, 3.5, 3.9, 4.2, 6]
plt.hist(values, bins = 3)
plt.show()
```

```python
# To customize a line graph
year = [1950, 1951, 1952, 2100]
pop = [2.538, 2.57, 2.62, 10.85]

# To add more data
year = [1800, 1850, 1900] + year
pop = [1.0, 1.262, 1.650] + pop

# To plot the data
plt.plot(year, pop)

# To edit the axes and title
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks([0, 2, 4, 6, 8, 10], ['0', '2B', '4B', '6B', '8B', '10B'])

# To display the plot
plt.show()
```
